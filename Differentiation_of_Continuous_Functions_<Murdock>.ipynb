# **Differentiation of continuous functions**

Recall the formulas for approximating derivatives:

Forward difference: $f'(x)\approx \frac{f(x+\Delta x)-f(x)}{\Delta x}$

Backward difference: $f'(x)\approx \frac{f(x)-f(x-\Delta x)}{\Delta x}$

Centered difference: $f'(x)\approx \frac{f(x_{i+1})-f(x_{i-1})}{2\Delta x}$

```python
# (1) The following code finds the forward, backward, and centered difference
# derivative approximations for a given function for a range of Delta x
# values, then plots the absolute errors versus Delta X on a log-log scale plot.

# "Magic" statement for jupyter notebook to display plots
%matplotlib inline
# Necessary imports
from numpy import *
from matplotlib.pyplot import *
## Initialization
# Point where derivative is evaluated
x = 0.5
#function to be differentiated
f = exp(x)
#Derivative of function (theo. calculation)
fPrime = exp(x)
#get vector of deltax values
deltax = 2.**arange(-1, -14, -1)
## Calculation
#Compute forward approx.
fPlus = exp(x + deltax)
forward = (fPlus - f)/deltax
#Compute backward approx
fMinus = exp(x - deltax)
backward = (f - fMinus)/deltax
#use average of forward and backward for centered
centered = (forward + backward)/2
#compute absolute errors
forwardError = abs(fPrime - forward)
backwardError = abs(fPrime - backward)
centerError = abs(fPrime - centered)
## Display section -- create figure
figure(1)
#plot forward, backward, and centered errors on log-log scale
plot(deltax, forwardError, label = "Forward Difference")
plot(deltax, backwardError, label = "Backward Difference")
plot(deltax, centerError, label = "Centered Difference")
# If limits are specified, set plot limits
xlim([1,1e-4])
ylim([1e-9,1])
# Use log-log scale
xscale("log")
yscale("log")
# Label plot and axes
xlabel("$\Delta x$")
ylabel("Absolute Error")
# Add legend and title
legend(loc = 3)
title("Errors for bw, fw, centered diff approx of $e^x$")
show()
```

```python
# (2) Using the graph from (1), estimate the slopes of these lines.
# Remember: change in log(absolute errro)/change in log(delta x)

# "Magic" statement for jupyter notebook to display plots
%matplotlib inline
# Necessary imports
from numpy import *
from matplotlib.pyplot import *
## Initialization
# Point where derivative is evaluated
x = 0.5
#function to be differentiated
f = exp(x)
#Derivative of function (theo. calculation)
fPrime = exp(x)
#get vector of deltax values
deltax = 2.**arange(-1, -14, -1)
## Calculation
#Compute forward approx.
fPlus = exp(x + deltax)
forward = (fPlus - f)/deltax
#Compute backward approx
fMinus = exp(x - deltax)
backward = (f - fMinus)/deltax
#use average of forward and backward for centered
centered = (forward + backward)/2
#compute absolute errors
forwardError = abs(fPrime - forward)
backwardError = abs(fPrime - backward)
centerError = abs(fPrime - centered)
## Display section -- create figure
figure(1)
# plot forward, backward, and centered errors on log-log scale
plot(log10(deltax), log10(forwardError), label="Forward Difference")
plot(log10(deltax), log10(backwardError), label="Backward Difference")
plot(log10(deltax), log10(centerError), label="Centered Difference")
# If limits are specified, set plot limits
xlim([0, -4])  # Corresponding to 1e-4 to 1
ylim([-9, 0])  # Corresponding to 1e-9 to 1
# Label plot and axes
xlabel("$\log(\Delta x)$")
ylabel("$\log(\mathrm{Absolute Error})$")
# Add legend and title
legend(loc=3)
title("Errors for bw, fw, centered diff approx of $e^x$")
# Show the plot
show()
# Calculate slopes
slope_forward = (log10(forwardError[1]) - log10(forwardError[2])) / (log10(deltax[1]) - log10(deltax[2]))
slope_backward = (log10(backwardError[1]) - log10(backwardError[2])) / (log10(deltax[1]) - log10(deltax[2]))
slope_centered = (log10(centerError[1]) - log10(centerError[2])) / (log10(deltax[1]) - log10(deltax[2]))
# Print the calculated slopes
print("Slope for Forward Difference:", slope_forward)
print("Slope for Backward Difference:", slope_backward)
print("Slope for Centered Difference:", slope_centered)
```

The slopes give you insight into the convergence behavior of the numerical differentiation methods used in your code.

**Forward Difference:**

* The slope values for the forward difference method are close to 1. This indicates that the absolute error is approximately proportional to the step size raised to the power of 1. It suggests a first-order convergence, which is expected for a forward difference.

**Backward Difference:**

* The slope values for the backward difference method are less than 1 but close to 1. This also suggests a first-order convergence, but it seems the convergence is slightly slower compared to the forward difference method.

**Centered Difference:**

* The slope values for the centered difference method are around 2. This indicates a second-order convergence, suggesting that the absolute error is proportional to the square of the step size. The centered difference method typically provides better accuracy compared to the forward and backward differences, and this is reflected in the faster convergence.


**Summary:**

* A ***slope of 1*** suggests first-order convergence.
* A ***slope of 2*** suggests second-order convergence.
* **The closer the slope is to the ideal value, the better the convergence behavior**.

```python
# (6a)
# Modifying the code from (1), compute (absolute error versus Delta x) graphs for
# estimators of the form a*FW + b*BW for three different combinations of a,b:
 # (i) a= 0.4, b = 0.6;
# (ii) a = 0.49, b = 0.51;
 # v(iii) a=0.499, b = 0.501.

%matplotlib inline
# Necessary imports
from numpy import *
from matplotlib.pyplot import *

## Initialization
# Point where derivative is evaluated
x = 0.5
# Function to be differentiated
f = exp(x)
# Derivative of function (theoretical calculation)
fPrime = exp(x)
# Get vector of deltax values
deltax = 2.**arange(-1, -14, -1)

# Define different combinations of a and b
combinations = [(0.4, 0.6), (0.49, 0.51), (0.499, 0.501)]

## Calculation and Display
# Create a figure
figure(1)

# Loop through different combinations of a and b
for a, b in combinations:
    # Compute forward and backward approximations
    fPlus = exp(x + deltax)
    forward = (fPlus - f) / deltax
    fMinus = exp(x - deltax)
    backward = (f - fMinus) / deltax

    # Compute estimator using the specified combination of a and b
    estimator = a * forward + b * backward

    # Compute absolute error
    error = abs(fPrime - estimator)

    # Plot the absolute error on log-log scale
    plot(deltax, error, label=f"{a}*Forward + {b}*Backward")

# If limits are specified, set plot limits
xlim([1, 1e-4])
ylim([1e-9, 1])
# Use log-log scale
xscale("log")
yscale("log")
# Label plot and axes
xlabel("$\Delta x$")
ylabel("Absolute Error")
# Add legend and title
legend(loc=3)
title("Errors for different estimators of the form $a*FW + b*BW$ for $e^x$")
show()

# Calculate slopes
slope_forward = (log10(forwardError[1]) - log10(forwardError[2])) / (log10(deltax[1]) - log10(deltax[2]))
slope_backward = (log10(backwardError[1]) - log10(backwardError[2])) / (log10(deltax[1]) - log10(deltax[2]))
slope_centered = (log10(centerError[1]) - log10(centerError[2])) / (log10(deltax[1]) - log10(deltax[2]))
# Print the calculated slopes
print("Slope for Forward Difference:", slope_forward)
print("Slope for Backward Difference:", slope_backward)
print("Slope for Centered Difference:", slope_centered)
```
